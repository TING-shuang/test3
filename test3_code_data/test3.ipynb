# Python函数、类与文件操作练习题

## 练习题（共20题）

### 函数基础

1. 定义一个函数`calculate_flow`，接收早高峰和晚高峰客流量参数，返回当日总客流量（示例调用：calculate_flow(85000, 72000)应返回157000）

2. 定义带返回值的函数`avg_speed`，计算3段路程的平均车速（输入参数：[60, 55, 70]，返回61.666...）

3. 嵌套函数练习：外层函数`bus_info`接收车牌号，内层函数`check_violation`判断是否超速（>80km/h），调用后输出"京A123 未超速"或"京B456 超速警告"

### 类与对象

4. 定义`BusLine`类，包含`line_num`（线路号）和`stations`（站点列表）属性，添加`add_station`方法用于添加新站点

5. 封装性练习：定义`MetroStation`类，使用私有属性`__daily_flow`存储日客流量，提供`set_flow`和`get_flow`方法

6. 单继承练习：父类`Vehicle`包含`plate`（车牌）属性，子类`ElectricBus`继承并添加`battery`（电量）属性

7. 多重继承练习：父类`Route`（线路号）和`Schedule`（发车间隔），子类`SubwayLine`继承两者并添加`peak_flow`（高峰客流量）属性

8. 多级继承练习：祖父类`Transport`（类型）→父类`Railway`（轨道类型）→子类`HighSpeedRail`（时速），要求子类能访问所有层级属性

9. 多态练习：父类`TrafficTool`有`move`方法（输出"移动中"），子类`Bus`重写为"公交行驶中"，子类`Bike`重写为"自行车骑行中"，创建对象并调用`move`方法

10. 方法类型练习：定义`TrafficData`类，包含实例方法`get_weekday_flow`（获取工作日客流）、类方法`update_version`（输出"数据版本V2.0"）、静态方法`format_time`（将秒转分钟，如300→5）

### 文件操作

11. 内置文件读取：使用`with open`读取`bus_data.txt`（内容：第一行"线路号:102"，第二行"站点数:15"），打印第二行内容

12. 内置文件写入：将公交线路信息["101路,12站", "102路,15站"]写入`new_bus.txt`，每行一条数据

13. Numpy文件读取：读取`passengers.csv`（5行2列数据：日期,客流量），使用`np.loadtxt`加载为字符串数组

14. Numpy文件写入：将客流量数组`[[10000, 12000], [9800, 11500]]`保存为`daily_flow.csv`，用逗号分隔

15. Pandas读取CSV：读取`metro_delay.csv`（包含"线路","延误次数"列），打印前3行数据

16. Pandas写入Excel：将处理后的公交线路数据（包含"线路号","高峰间隔"列）保存为`bus_schedule.xlsx`，工作表命名"高峰表"

17. 混合读取：先用`readline`读取`traffic_log.txt`第一行日志类型，再用`readlines`读取剩余日志内容

18. 二进制文件操作：使用`np.save`保存地铁站点坐标数组`[[39.9, 116.4], [39.8, 116.5]]`为`station_coords.npy`，再用`np.load`读取验证

19. 异常处理（选做）：在文件读取时添加`try-except`，当文件不存在时输出"文件未找到，请检查路径"

20. 综合应用：创建`TrafficAnalyzer`类，包含`load_data`方法（用Pandas读取CSV）和`calculate_avg`方法（计算平均客流量），实例化后调用方法处理`month_flow.csv`数据

## 参考答案（部分示例）

```python
# 第1题参考答案
def calculate_flow(morning_peak, evening_peak):
    return morning_peak + evening_peak

print(calculate_flow(85000, 72000))
```

```python
# 第2题参考答案
def avg_speed(speeds):
    return sum(speeds) / len(speeds)

print(avg_speed([60, 55, 70]))
```

```python
# 第3题参考答案
def bus_info(plate, speed):
    def check_violation(current_speed):
        return "超速警告" if current_speed > 80 else "未超速"

    print(f"{plate} {check_violation(speed)}")

bus_info("京A123", 60)
bus_info("京B456", 90)
```

```python
# 第4题参考答案
class BusLine:
    def __init__(self, line_num, stations):
        self.line_num = line_num
        self.stations = stations

    def add_station(self, station):
        self.stations.append(station)

line = BusLine("102路", ["A站", "B站"])
line.add_station("C站")
print(line.line_num, line.stations)
```

```python
# 第5题参考答案
class MetroStation:
    def __init__(self, daily_flow=0):
        self.__daily_flow = daily_flow

    def set_flow(self, daily_flow):
        self.__daily_flow = daily_flow

    def get_flow(self):
        return self.__daily_flow

station = MetroStation()
station.set_flow(120000)
print(station.get_flow())
```

```python
# 第6题参考答案
class Vehicle:
    def __init__(self, plate):
        self.plate = plate

class ElectricBus(Vehicle):
    def __init__(self, plate, battery):
        super().__init__(plate)
        self.battery = battery

electric_bus = ElectricBus("京A123", "85%")
print(electric_bus.plate, electric_bus.battery)
```

```python
# 第7题参考答案
class Route:
    def __init__(self, line_num):
        self.line_num = line_num

class Schedule:
    def __init__(self, interval):
        self.interval = interval

class SubwayLine(Route, Schedule):
    def __init__(self, line_num, interval, peak_flow):
        Route.__init__(self, line_num)
        Schedule.__init__(self, interval)
        self.peak_flow = peak_flow

subway = SubwayLine("1号线", "5分钟", 180000)
print(subway.line_num, subway.interval, subway.peak_flow)
```

```python
# 第8题参考答案
class Transport:
    def __init__(self, transport_type):
        self.transport_type = transport_type

class Railway(Transport):
    def __init__(self, transport_type, railway_type):
        super().__init__(transport_type)
        self.railway_type = railway_type

class HighSpeedRail(Railway):
    def __init__(self, transport_type, railway_type, speed):
        super().__init__(transport_type, railway_type)
        self.speed = speed

hsr = HighSpeedRail("铁路交通", "高速铁路", 350)
print(hsr.transport_type, hsr.railway_type, hsr.speed)
```

```python
# 第9题参考答案
class TrafficTool:
    def move(self):
        print("移动中")

class Bus(TrafficTool):
    def move(self):
        print("公交行驶中")

class Bike(TrafficTool):
    def move(self):
        print("自行车骑行中")

for tool in [TrafficTool(), Bus(), Bike()]:
    tool.move()
```

```python
# 第10题参考答案
class TrafficData:
    version = "V1.0"

    def __init__(self, weekday_flow):
        self.weekday_flow = weekday_flow

    def get_weekday_flow(self):
        return self.weekday_flow

    @classmethod
    def update_version(cls):
        cls.version = "V2.0"
        print(f"数据版本{cls.version}")

    @staticmethod
    def format_time(seconds):
        return seconds // 60

traffic_data = TrafficData(150000)
print(traffic_data.get_weekday_flow())
TrafficData.update_version()
print(TrafficData.format_time(300))
```

```python
# 第20题参考答案
import pandas as pd

class TrafficAnalyzer:
    def __init__(self, file_path):
        self.file_path = file_path
        self.data = None

    def load_data(self):
        self.data = pd.read_csv(self.file_path)
        return self.data

    def calculate_avg(self):
        return self.data["客流量"].mean()

analyzer = TrafficAnalyzer("month_flow.csv")
analyzer.load_data()
print(analyzer.calculate_avg())
```
